```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_latent = pd.read_csv('AttentionMean.csv', header=None)
```

```python
mp_latent=df_latent.to_numpy()
```

```python
mp_latent.shape
```

```python
# Step 1: Generate synthetic latent space data (e.g., 100 samples with 50 features)
latent_space_data = np.random.randn(100, 50)

# Step 2: Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
latent_space_2d = tsne.fit_transform(mp_latent)

# Step 3: Plot the t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(latent_space_2d[:, 0], latent_space_2d[:, 1])
plt.title("t-SNE plot of Latent Space")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```

```python
latent_space_data.shape
```

```python
df1 = pd.DataFrame(latent_space_2d)
df1.columns = ['Feature1', 'Feature2']
```

```python
df1['y'] = df_latent.iloc[:,20]
```

```python
df1.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Feature1', y='Feature2',
    hue='y',
    palette=sns.color_palette("hsv", len(np.unique(df1['y']))),
    data=df1,
    legend="full",
    alpha=0.7
)

plt.title('t-SNE Visualization of Clusters')
plt.savefig('latentattentionSpace_for_device15.png')
plt.show()
```
